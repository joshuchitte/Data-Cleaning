Data Cleaning

```python
df.isnull().sum()
```

```python
int(df["Age"].mean())
```

```python
# replace the missing values with median
df['Age']=df['Age'].fillna(np.mean(df['Age']))
```

```python
df.isnull().sum()
```

```python
df['Embarked'].mode()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Cabin'].value_counts()
```

```python
df['Cabin'].mode()
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
```
